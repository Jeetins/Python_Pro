# Dictionary
1. Dictionary is a mutable data type in Python.
2. A python dictionary is a collection of key and value pairs separated by a colon (:) & enclosed in curly braces {}.
3. Keys must be unique in a dictionary, duplicate values are allowed.



```python
d={0:1} # In Python, dictionaries are defined within braces {} with each item being key:value
type(d)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False} # 	Value can be of any data type
print(j)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}} # 	If at least one value is dict, then it is called as nested Dictionary
print(j)
```

## Dict.items()
1. Items() methods returns a list of key-value pairs in the dictionary
2. List of tuples
3. Each tuples consist of (key, value)


```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}} # Access each key-value pair within a dictionary
j.items()
```

## Dict.keys()
- Keys() methods return a list of keys in the dictionary. 


```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}}
j.keys()
```

## Dict.values()
- values() methods return a list of values in the dictionary. 


```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}}
j.values()
```

## Adding Single Item in Dictionary. 

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False} # Updating single Item to Dictionary. 
j.update({"car23":{1:2}})
print(j)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False} # Updating single Item to Dictionary. 
j["car23"]={1:2}
print(j)
```

```python
j={0:1,2.3:9.6,True:[1,2],} # Updating Multiple Item to Dictionary. 
j.update({"car23":{1:2},'srk':{1,2,3},(1,2):False})
print(j)
```

```python
keys = {'a' , 'b' , 'c' , 'd'}
value = [10,20,30]
mydict3 = dict.fromkeys(keys , value) # Create a dictionary from a sequence
mydict3
```

```python
value.append(40) 
mydict3
```

## Deleting Item in Dictionary.

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}}
del j[True]
print(j)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}} # Pop() method removes and returns the value of the specified key. 
j.pop('srk') # If the key doesnt exist, it raises a key error 
print(j)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3},(1,2):False,"car23":{1:2}}
j.clear()
print(j)
```

## Dict.copy()
1. Copy() method returns a shallow copy of the Dictionary
2. Changes to the copy of the dictionary, dont affect the original dictionary. 


```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3}}
k=j.copy()
k.update({(1,2):False,"car23":{1:2}})
print(j)
print(k)
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3}}
k=j.copy()
k.update({(1,2):False,"car23":{1:2}})
print(j)
print(k)
```

## Replace Item in Dictionary.

```python
mydict1 = {'Name':'Asif' , 'ID': 12345 , 'DOB': 1991 , 'Address' : 'Hilsinki'} #  We can create same Key which will replace the existing Value. 
mydict1
```

```python
mydict1['DOB'] = 1992 # Changing Dictionary Items
mydict1['Address'] = 'Delhi'
mydict1
```

```python
j={0:1,2.3:9.6,True:[1,2],'srk':{1,2,3}} #  We can also remove the Key fist by using pop and replace it with Other key.
j[True]=j.pop('srk')
print(j)
```

```python
mydict1['Job'] = 'Analyst' # Adding items in the dictionary
mydict1
```

```python
mydict1.pop('Job') # Removing items in the dictionary using Pop method
mydict1
```

```python
del[mydict1['ID']] # Removing item using del method
mydict1
```

```python
del mydict1 # Delete the dictionary object
mydict1
```
